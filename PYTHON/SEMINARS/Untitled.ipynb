```python
# 1.Определить корни
# 2.Найти интервалы, на которых функция возрастает
# 3.Найти интервалы, на которых функция убывает
# 4.Построить график
# 5.Вычислить вершину
# 6.Определить промежутки, на котором f > 0
# 7.Определить промежутки, на котором f < 0

# import numpy as np
# from sympy import *
# from sympy.plotting import plot
import matplotlib.pyplot as plt
import numpy as np

# x = Symbol('x', real=True)
# f = -12*x**4*sin(cos(x)) - 18*x**3+5*x**2 + 10*x - 30

def f(x):
    return -12*x**4*np.sin(np.cos(x)) - 18*x**3 + 5*x**2 + 10*x - 30
```

```python
# 1. определить корни
# nsolve(f, x, 0)
roots = np.roots([-12, -18, 5, 10, -30])
print(roots)
```

```python
# функция возрастает

# solveset(f>0, x, S.Reals)   

def find_intervals(f, a, b, step):
  x = a
  while x <= b:
    y1 = f(x)
    y2 = f(x + step)
    if y1 < y2:
      print(f"Функция возрастает на интервале [{x}, {x + step}]")
    x += step

find_intervals(f, -100, 100, 10)
```

```python
#  функция убывает

# solveset(f<0, x, S.Reals)

def find_intervals(f, a, b, step):
  x = a
  while x <= b:
    y1 = f(x)
    y2 = f(x + step)
    if y1 > y2:
      print(f"Функция убывает на интервале [{x}, {x + step}]")
    x += step

find_intervals(f, -100, 100, 10)
```

```python
x_start = -100
x_end = 100

# Создаем массив с значениями x на заданном интервале
x_values = np.linspace(x_start, x_end, 50)

plt.plot(x_values, f(x_values))
plt.grid()
plt.xlabel('x')
plt.ylabel('y')
```
